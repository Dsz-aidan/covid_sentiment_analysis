```python
import pandas as pd

# Load your dataset. Replace 'your_dataset.csv' with your dataset file.
data = pd.read_csv('covid19_tweets.csv')
data.head()

# Define a list of patterns to search for
usa_patterns = ['usa', 'u\\.s\\.a\\.', 'united states', 'america']

# Create a regex pattern by joining the list with '|'
usa_regex = '|'.join(usa_patterns)

# Filter data using regex pattern
usa_data = data[data['user_location'].str.contains(usa_regex, case=False, na=False)]

# Select relevant columns
usa = usa_data[['text','user_location']]

usa
```

```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize the VADER sentiment analyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

# Function to determine sentiment and map to 0, 1, or 2
def map_sentiment(text):
    sentiment_scores = sia.polarity_scores(text)

    if sentiment_scores['pos'] > sentiment_scores['neg']:
        return 1  # Positive
    elif sentiment_scores['neg'] > sentiment_scores['pos']:
        return 0  # Negative
    else:
        return 2  # Neutral

# Apply sentiment analysis to the 'text' column and create a 'target' column
usa['target'] = usa['text'].apply(map_sentiment)

# Combine all columns into a single DataFrame
final_df = usa[['text','user_location','target']]

# Specify the CSV file name
csv_file = "final.csv"

# Save the final DataFrame to a CSV file
final_df.to_csv(csv_file, index=False)

print(f"Final DataFrame has been saved to {csv_file}")
```
